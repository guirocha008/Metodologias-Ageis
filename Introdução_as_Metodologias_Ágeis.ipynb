# $$Introdução~as~Metodologias ~Ágeis$$

### $$CONCEITO:$$

-----------------------------------------------------------------------------------------------------------------------


**Os métodos ágeis surgiram para resolver problemas comuns a quase toda organização que precisa gerenciar projetos**


**Uma organização nasce para resolver um problema, normalmente a missão da empresa reflete esse problema**

-----------------------------------------------------------------------------------------------------------------------


## $$Os~ três~ tipos ~de ~problemas ~organizacionais$$

-----------------------------------------------------------------------------------------------------------------------

- **Conhecidos:** Várias soluções possíveis para resolver o problema, referência clara no passado


- **Conhecíveis:** O problema já existe e não existe uma solução para ele no passado, ou ele já existia no passado e as soluções até hoje existentes não servem mais para esse problema, os dados para problemas conhecíveis estão no presente  


- **Incognoscíveis:** Um problema que está no futuro, que pode acontecer ou não acontecer e não há nenhuma solução para o problema ainda, dados estão no futuro


-----------------------------------------------------------------------------------------------------------------------

### $$Métodos~ Preditivos~ vs ~Adaptativos(Agil)~ vs~ Sinais~ Fracos$$

-----------------------------------------------------------------------------------------------------------------------

**Métodos Preditivos: Problemas Conhecidos**

- **Preditivos PDCA(Plan, Do, Check, Act):** 
   - Planejar, fazer, checar e agir
   - Ciclo de Deming, os dados são fornecidos Apriori
   - Escopo conhecido Apriori, os requisitos são detalhados no começo do projeto: Custo e Prazo


**Métodos Adaptativos(Ágil): Problema Conhecíveis**

- **Abordagem DCAP(Do, Check, Act, Plan):** 
   - Fazer, Checar, Agir e Planejar
   - Os dados para tomar decisão vem da execução
   - Escopo nasce a partir da execução, Tolerância zero é criada parar prazo e custos


**Métodos Sinais Fracos: Problemas Incognoscíveis**
- **Abordagem híbrida:**
    - O objetivo do estudo de sinais fracos é antecipar-se a eventos incertos, inesperados, com potencial de impacto significativo, visando preparar os gestores para decidir ou agir quando os mesmos ocorrerem.
    - Os dados não se sabem
    - Escopo ainda não se conhece

-----------------------------------------------------------------------------------------------------------------------


### $$Técnica~ Lean ~Startup:$$

-----------------------------------------------------------------------------------------------------------------------

**Esse método ensina como criar produtos desejados por clientes, gerando ciclos de aprendizado rápidos, em que as mudanças no direcionamento das estratégias da empresa aconteçam visando um crescimento acelerado**

**Este princípio(Construir, Medir e Aprender) é essencial para a compreensão do que é Lean Startup, pois define resumidamente o caminho a ser seguido para transformar ideias em negócios lucrativos. Transformar as ideias em produtos e medir a reação dos clientes são guias fundamentais a este princípio**

- **Construir:** Construir rapidamente um produto minimamente viável

- **Medir:** Medir sua eficácia no mercado

- **Aprender:** Aprender com essa experiência

-----------------------------------------------------------------------------------------------------------------------

## $$Modelo~ CYNEFIN:~ Tipos~ de ~decisão$$

-----------------------------------------------------------------------------------------------------------------------

**Domínio do Conhecido:**

- **Claro(Melhores Práticas):**
  - Problema do passado, uma única forma de resolver 

  - ECR(Entender, Categorizar e Resolver)

  - Medir através da qualidade e eficiência

  - Liderança: estilo formal, controle centralizado

  - Número de pessoas que podem trabalhar no problema: >150


- **Complicado(Boas Práticas):**
  - Problema do passado

  - Várias formas de resolver

  - EAR(Entender, Analisar e Resolver)

  - Medir através da Eficácia, liderança: Especialista, integradora

  - Número de pessoas que podem trabalhar no problema: <150


**Domínio do Conhecível:**


- **Complexo(Praticas Emergentes):**
  - Problema do presente

  - SER(Sondar, Entender e Resolver)

  - Medir através do aprendizado

  - liderança: Equipe(informal ou facilitadora)

  - Número de pessoas que podem trabalhar no problema: 15


- **Caos(Praticas Inéditas):**
  - Problema do presente

  - AER(Agir, Entender e Resolver)

  - Medir através dos dados evitados

  - Liderança: Não existe

  - Número de pessoas que podem trabalhar no problema: 5


**Domínio Central:**


- **Confuso(Mais de 1 tipo de decisão)**

## $$As~ 3 ~fases ~de~ uma ~startup$$

-----------------------------------------------------------------------------------------------------------------------




### **Problem Solution Fit: Há uma solução conhecida e aplicavem para o problema?**


- **1° Exploração do Problema:**  
   - **Solução:** P&D, Mapa do Problema
   
   
- **2° Geração de ideia:** É possível ter soluções para resolver? 
   - **Solução:** Design Thinking
   
   
- **3° Protótipo:** Testar novas ideias? 
   - **Solução:** Prototipagem, Storyboarding
   
-----------------------------------------------------------------------------------------------------------------------


### **1° Abismo: Entre Solution Fite e Market Fit**


### **Product Market Fit: O mercado está respondendo ao produto com estusiasmo e a difusão está aumentando?**



- **4° MVP:** Mercado vai aceitar minha solução? Aqui os primeiros adeptos a aceitarem a ideia 
   - **Solução:** Lean Startup


- **5° Versão beta:** Os primeiros a seguir os primeiros adeptos, seguidores de quem aceitou o produto 
   - **Solução:** Scrum


- **6° Versão Candidata:** Validar com a maioria dos adeptos iniciais, caso essa etapa se confirme, poderá passar para o 2 abismo 
   - **Solução:** Agile GP
   
   
-----------------------------------------------------------------------------------------------------------------------


### **2° Abismo: Entre Market Fit e Scale**

### **Scale: O Produto pode ser disponibilizado ao mercado de uma forma plena e sustentável?**


- **7° Versão Final:** Lançar a versão final do produto 
   - **Solução:** GP - Híbrido


- **8° Consolidação no Mercado:** Tornar a versão final cada vez mais estável, repetitiva e enxuta 
   - **Solução:** Lean Thinking
   

- **9° Escala:** Tornar padrão de mercado 
   - **Solução:** 6 Sigma

## **Lean(Enxuto):** 

**Proveniente do Lean manufacturing, Sistema Toyota de produção. (JIT) Just in Time**

**4 Pilares do sistema Toyota:**

- **1º Pilar - Philosophy:** Decisões Gerenciais baseadas no longo prazo, o que realmente a organização quer resolver


- **2º Pilar - Process:** Criar processos para que os produtos circulem ao longo do processo, e não fiquem emperrados


- **3º Pilar - People and Partners:** O que fazemos na empresa, aplicamos fora do trabalho também. Respeitar, Desenvolver e Desafiar. **Estado de flow** = Ocorre quando uma pessoa realiza uma atividade em imersão, sentindo maior energia, prazer e foco durante a execução.


- **4º Pilar - Problem Solving:** Metodologia Kaizen, aprendizado continuo na organização, tomar decisões com cautela, demore para tomar uma decisão no momento que decidir implemente rapidamente sua decisão





**Os sete desperdícios do Lean:**


- **Superprodução:** Fazer mais itens do que necessita

- **Defeitos:** Quantidade de defeitos, reclamações e perdas em decorrência disso

- **Transporte:** Movimentação desnecessária do produto 

- **Espera:** O quanto uma área de negócio espera outra para começar

- **Inventário:** Quantidade correta no estoque?, mais ou menos

- **Movimentação:** Os funcionários movimentando mais que o necessário para poder fazer o trabalho

- **Processamento Excessivo:** Fazer mais trabalho em um produto do que o necessário

## **Agile(Adaptative):**


**Um grupo de desenvolvedores de software percebeu que a forma de resolver problemas e desenvolver produtos precisava ser diferente e, juntos, criaram o Manifesto Ágil:**


- **Focar em Indivíduos e Interações:** Não em Processos e Ferramentas


- **Focar em Software Funcionando:** Não em Documentação Extensa


- **Focar em Colaboração com o Cliente:** Não em Negociação de Contratos


- **Focar em Responder a Mudanças:** Não em Seguir um Plano


**CONCEITOS DO AGILE:**
 
- **1° Visão:** Antes de iniciar o projeto, Aborda os problemas principais que o produto deseja resolver, é a Visão do Produto que trará os insumos necessários para criar e manter o Backlog do Produto
   - Qual o valor do produto?
   - Por que estamos fazendo isso?


- **2° Roadmap:** Que caminho vamos percorrer para alcançarmos a Visão?
   - **Data** = Quando estará disponível? 
   - **Nome** = Qual o nome da versão? 
   - **Meta** = Por que estamos fazendo isso? 
   - **Requisitos** = Quais são as principais funções
   - **Métricas** Como saberemos que a meta foi alcançada?



- **3° Release:** De quanto em quanto tempo iremos ao mercado em busca de feedback?
   - Teste do produto com o usuário


- **4° Sprint:** Como, incrementalmente, construiremos nosso lançamento, Como dividiremos o trabalho até o próximo lançamento?
   - Teste do produto com a equipe


- **5° Daily:** Diariamente, como acompanharemos nosso progresso.

## **Scrum:** 

**O Scrum é uma estrutura que ajuda as equipes a trabalharem juntas. Semelhante a uma equipe de rugby (de onde vem o nome) treinando para o grande jogo, o Scrum estimula as equipes a aprenderem com as experiências, a se organizarem enquanto resolvem um problema e a refletirem sobre os êxitos e fracassos para melhorarem sempre**

### Pilares do Scrum:


- **TRANSPARÊNCIA** Tanto quem estiver fazendo quanto quem estiver recebendo o trabalho deve ter uma visão clara sobre todo o processo


- **INSPEÇÃO** Todos os detalhes referentes ao processo do scrum devem ser inspecionados com frequência para identificar possíveis problemas e corrigi-los em tempo hábil


- **ADAPTAÇÃO** A análise constante do scrum permitirá encontrar oportunidades de ajustes que devem ser feitos rapidamente, o que irá exigir um alto potencial de adaptação.



### Praticas Fundamentais para usar o Scrum: Papeís, Eventos e Artefatos:


### **Papeís:**

- **Product Owner:** Dono do Produto, responsável por criar a visão do produto, foco prioridades para o negócio, adição de valor


- **SCRUM Master:** Líder servidor, facilitador, removedor de impedimentos do Time Scrum e demais stakeholders, foco nos processos e no desenvolvimento das pessoas


- **Desenvolvedores:** Formado por equipe auto-organizada, responsável por realizar o trabalho que irá gerar a entrega da Sprint, foco nas entregas, valor adicionado em uso


### **Eventos:**

- **Sprint:**

  - **O que é:** Sprint é o coração do Scrum. Onde um incremento de produto, potencialmente utilizável é criado.
  - **Duração:** No máximo 4 semanas.
  - **Quem participa:** Time Scrum.
  
  
- **Sprint Planning:**

   - **O que é:** Reunião colaborativa do Time Scrum, onde é definido pelo Time de Desenvolvimento, os itens de Backlog do Produto que serão entregues ao final da Sprint. Estes itens selecionados pelo Time de Desenvolvimento são chamados da Sprint Backlog. Durante esta etapa de planejamento, é definido também a meta da sprint, ou seja, o objetivo definido para a Sprint que pode ser satisfeito através da implementação dos itens da Sprint Backlog.
   - **Duração:** No máximo 8 horas.
   - **Quem participa:** Time Scrum.
  
  
  
- **Daily Scrum:**

   - **O que é:** Reunião diária, executada pelo Time de Desenvolvimento e Scrum Master, onde objetiva-se o sincronismo das tarefas e planejamento para as próximas 24h. Algumas questões importantes devem ser levadas para a Daily Scrum, como por exemplo: O que foi feito no dia anterior, o que será feito no dia atual e possíveis impedimentos que estão dificultando a conclusão de tarefas.
   - **Duração:** No máximo 15 minutos.
   - **Quem participa:** Time de Desenvolvimento e Scrum Master.
   
   
- **Sprint Review:**

   - **O que é:** Reunião realizada no último dia da Sprint, onde é demonstrado para o product Owner e demais stakeholders, o incremento do produto criado. Focado no aceite ou rejeição do trabalho realizado. O resultado da Revisão da Sprint é um Backlog de Produto revisado, que define os itens prováveis de Backlog da próxima Sprint.
   - **Duração:** No máximo 4 horas.
   - **Quem participa:** Time Scrum e demais stakeholders que o Product Owner julgar necessário.
   
   
- **Sprint Retrospective:**

   - **O que é:** Reunião que ocorre após a Revisão da Sprint. Focado em rever o processo de trabalho e o que poderá ser melhorado. Está relacionada a garantia da qualidade.
   - **Duração:** No máximo 3 horas.
   - **Quem participa:** Time Scrum
  





### **Artefatos:**

- **Product Backlog:**
   - Criado a partir da Visão do Produto
   - É uma lista ordenada de tudo aquilo que é necessário para criar o produto. 
   - O que estiver fora desta lista não será feito. 
   - O Backlog do Produto contém requisitos de negócio, não funcionais, melhorias, correções, testes de arquitetura, etc

- **Sprint Backlog:**

  - Itens do Product Backlog que foram selecionados para a Sprint. O Backlog da Sprint retrata os itens que estão sendo trabalhados e que estarão disponíveis no próximo incremento.

- **Definition of Done:**

  - Ao definir que um item está pronto, é preciso que todos tenham o mesmo conceito do que é pronto na determinada Sprint. O Time deve ter um entendimento compartilhado do que significa o trabalho estar completo, assegurando a transparência.
